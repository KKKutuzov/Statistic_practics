```python
def ff(x):
    return sts.norm().rvs(1) * x**4 - 3 * x**2 + x + 1


vfunc = np.vectorize(ff)
k = 5
for m in [10, 20, 30]:
    plt.figure(figsize=(20, 20))
    ax = plt.subplot(3, 1, m / 10)
    k = 4
    x = np.zeros((2 * m))
    X = np.zeros((k, 2 * m))
    for i in range(2 * m):
        x[i] = i / m
    for i in range(k):
        X[i] = x ** (i + 1)
    X = X.transpose()
    XXX = X
    z = 1.2
    y = np.dot(X, np.array([4, 0, -3, 1])) + 1 + sts.norm().rvs(2 * m)
    reg = LinearRegression().fit(X, y)
    reg.score(X, y)
    # print('a = ', reg.coef_, 'b = ' , reg.intercept_)
    ax.scatter(x, y, c="deeppink")
    plt.plot(
        x,
        np.dot(X, reg.coef_) + reg.intercept_,
        label="Regres d = {}".format(k),
        alpha=0.5,
    )
    print(
        "  x=1.2  d = {} : y = ".format(k),
        np.dot([z**4, z**3, z**2, z], reg.coef_) + reg.intercept_,
    )
    ar = x ** (k + 1)
    ar = ar.reshape(-1, 1)
    X = np.hstack((X, ar))
    reg = LinearRegression().fit(X, y)
    reg.score(X, y)
    # print('a = ', reg.coef_, 'b = ' , reg.intercept_)
    plt.plot(
        x,
        np.dot(X, reg.coef_) + reg.intercept_,
        label="Regres d = {}".format(k + 1),
        alpha=0.5,
    )
    print(
        "  x=1.2  d = {} : y = ".format(k + 1),
        np.dot([z**5, z**4, z**3, z**2, z], reg.coef_) + reg.intercept_,
    )
    ar = x ** (k + 2)
    ar = ar.reshape(-1, 1)
    X = np.hstack((X, ar))
    ar = x ** (k + 3)
    ar = ar.reshape(-1, 1)
    X = np.hstack((X, ar))
    reg = LinearRegression().fit(X, y)
    reg.score(X, y)
    # print('a = ', reg.coef_, 'b = ' , reg.intercept_)
    plt.plot(
        x,
        np.dot(X, reg.coef_) + reg.intercept_,
        label="Regres d = {}".format(k + 3),
        alpha=0.5,
    )
    print(
        "  x=1.2  d = {} : y = ".format(k + 3),
        np.dot([z**7, z**6, z**5, z**4, z**3, z**2, z], reg.coef_)
        + reg.intercept_,
    )
    y = np.dot(XXX, np.array([4, 0, -3, 1])) + 1
    plt.plot(XXX[:, 0], y, label="", alpha=0.5)
    plt.legend(loc="best")


plt.show()
```
