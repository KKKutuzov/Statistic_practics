```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from sklearn.linear_model import LinearRegression

%matplotlib inline
from IPython.display import Image

display(Image(filename="task_12.png", width=800))
```

Генерируем данные по закону:

$$Y_i = aX_i + b + \varepsilon_i\text{, где } \varepsilon \sim N(0,1)$$

```python
a, b = 2, 3
for m in [10, 20, 30]:
    plt.figure(figsize=(10, 10))
    ax = plt.subplot(3, 1, m / 10)
    x = np.zeros((m, 1))
    for i in range(m):
        x[i] = i / m
    y = np.dot(x, np.array([a])) + b + sts.norm().rvs(m)
    z1 = []
    for j in range(m):
        for i in range(j):
            z1.append((y[i] - y[j]) / (x[i] - x[j])[0])
    z2 = []
    for j in range(m):
        z2.append(y[j] - np.median(z1) * x[j][0])
    ax.scatter(x, y, c="deeppink")
    plt.plot(
        x,
        np.dot(x, np.median(z1)) + np.median(z2),
        label="regression new method",
        alpha=0.5,
    )
    plt.plot(x, np.dot(x, a) + b, label="true line", alpha=0.5)
    print("Оценка на a = {}, оценка на b = {}".format(np.median(z1), np.median(z2)))
    print("RSS = ", np.mean((y - x[:, 0] * np.median(z1) - np.median(z2)) ** 2))
    plt.legend(loc="best")
    plt.show()
```

Построим теперь на одном графике и прямую полученную методом МНК и методом Тейла

```python
a, b = 2, 3
for m in [10, 20, 30]:
    plt.figure(figsize=(10, 10))
    ax = plt.subplot(3, 1, m / 10)
    x = np.zeros((m, 1))
    for i in range(m):
        x[i] = i / m
    y = np.dot(x, np.array([a])) + b + sts.norm().rvs(m)
    z1 = []
    for j in range(m):
        for i in range(j):
            z1.append((y[i] - y[j]) / (x[i] - x[j])[0])
    z2 = []
    for j in range(m):
        z2.append(y[j] - np.median(z1) * x[j][0])
    ax.scatter(x, y, c="deeppink")
    plt.plot(
        x,
        np.dot(x, np.median(z1)) + np.median(z2),
        label="regression new method",
        alpha=0.5,
    )
    plt.plot(x, np.dot(x, a) + b, label="true line", alpha=0.5)
    print("Метод Тейла:")
    print("Оценка на a = {}, оценка на b = {}".format(np.median(z1), np.median(z2)))
    print("RSS = ", np.mean((y - x[:, 0] * np.median(z1) - np.median(z2)) ** 2))
    print("\nМНК:")
    reg = LinearRegression().fit(x, y)
    print("Оценка на a = ", reg.coef_[0], ", оценка на bb = ", reg.intercept_)
    print("RSS = ", np.mean((y - x[:, 0] * reg.coef_[0] - reg.intercept_) ** 2))
    plt.plot(x, np.dot(x, reg.coef_) + reg.intercept_, label="Regres", alpha=0.5)

    plt.legend(loc="best")
    plt.show()
```
