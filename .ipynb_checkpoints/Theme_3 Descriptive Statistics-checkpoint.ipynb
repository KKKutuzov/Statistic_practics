## 3.2.2. Для массива данных "titanic"

### 1) Построить диаграмму рассеяния для age и fare, цветом задать пол, формой точек – класс;

```python
plt.figure(figsize=(10, 8))

sb.scatterplot(x='fare', y='age',hue='sex', style='pclass', data=df)

plt.show()
```

### 2) Cравнить выживших и невыживших пассажиров с помощью параллельных координат, используя столбцы survived, pclass, who, age, fare.

```python
plt.figure(figsize=(6, 5)) 

pd.plotting.parallel_coordinates(
    df[['survived',  'pclass',  'age', 'fare']].assign(Woman=(df.who=='woman')*150),
    class_column='survived', 
    color=('#C7F464', '#556270')
)

plt.show()
```

### 3) Построить boxplot() для fare по классам.

```python
sb.boxplot(x="pclass", y="fare", data=df[df['fare']<300])

plt.show()
```

### 3.3.2. 1) Моделировать 1000 реализация с.в. X ∼ Cauchy, и с.в. Y ∼ N (0, 1). Построить гистограммы отдельно и на одном графике, сравнить.

```python
X_sample = sts.cauchy().rvs(1000)
Y_sample = sts.norm().rvs(1000)
plt.hist(X_sample,bins = 30, label = 'X_sample hist')
plt.show()
plt.hist(Y_sample, bins = 30, label = 'Y_sample hist')
plt.show()
plt.hist(X_sample, bins = 30, label = 'X_sample hist')
plt.hist(Y_sample, bins = 30, label = 'Y_sample hist')
plt.show()
```

## 2) Независимые с.в. X, Y ∼ N (0, 1). Моделировать 1000 реализация с.в.(X + Y, X − Y ), построить ядерную оценку двумерной плотности.

```python
X_sample = sts.norm().rvs(1000)
Y_sample = sts.norm().rvs(1000)
x = X_sample + Y_sample
y = X_sample - Y_sample
sb.kdeplot(x, y)
plt.show()
```
